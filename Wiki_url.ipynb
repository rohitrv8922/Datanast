**Need to import all important libraries**

```python
import pandas as pd
import requests
# BeautifulSoup is imported with the name bas4 
import bs4
import json
```

**I have taken the url of english wifipedia**

```python
URL = 'https://en.wikipedia.org/wiki/Main_Page'
```

**Now fetching all the html source from the url**

```python
# Fetch all the HTML source from the url
response = requests.get(URL)


soup = bs4.BeautifulSoup(response.text, 'html.parser')
links = soup.select('a')
```

**Now run the loop to fetch all the url in wikipedia fro  top to bottom one by one**

```python

# Print out the result
for link in links:
  print(link.get_text())
  if link.get('href') != None:
    if 'https://' in link.get('href'):
      print(link.get('href'))
    else:
      print('https://en.wikipedia.org' + link.get('href')) # Convert relative URL to absolute URL

  print('----------------------------') # Just a line break
```

```python
data = pd.DataFrame({
    "Url": links
})
```

```python
data.to_csv('file.csv', index=False)
```
